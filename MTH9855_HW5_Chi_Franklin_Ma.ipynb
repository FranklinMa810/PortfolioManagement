```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
mu, sigma = 0, 1.0
X = np.random.normal(mu, sigma, (500,10))
print X.shape
epi = np.random.normal(mu,sigma, (500,1))
Y = 3 * X[:,1] + 17 * X[:,2] + epi[:,0]
print Y.shape
```

```python
# We do not consider Cross validation and training and test
coef = []
n = 100  # lambda numbers    
lam_start = 10
lam_end = -2
lams = np.linspace(lam_start, lam_end, n)
for lam in lams:
    clf = linear_model.Lasso(lam)
    clf.fit(X, Y)
    coef.append(clf.coef_)
```

```python
df = pd.DataFrame(coef)
df.columns = ["X1","X2","X3","X4","X5","X6","X7","X8","X9","X10"]
df["lambda"] = lams
print(df.shape)
df.head()
```

```python
names = ["X1","X4","X5","X6","X7","X8","X9","X10"]
feature = df
for name in names:
    feature = feature[feature[name] == 0.0]
```

```python
names2 = ["X2", "X3"]
for name in names2:
    feature = feature[feature[name] != 0.0]
```
