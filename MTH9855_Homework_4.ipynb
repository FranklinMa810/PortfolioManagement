#### Problem 4.2

Write a function which takes an input data frame, and estimates the
volatilities and correlations that you will need. Specifically, let $t$ be the date of the
last row in the input. As discussed in Lecture, estimate volatilities and correlations
separately. Let $S$ be a diagonal matrix with the vols on the diagonal, and let $R$ be
the correlation matrix, so that the covariance matrix is 
$$
\Sigma = SRS.
$$ 
Use a shrinkage
estimator, specifically,
$$
R \leftarrow \text{shrinkage}\times R + (1 - \text{shrinkage}) \times \mathbb{1}
$$
where shrinkage is a parameter, and $\mathbb{1}$ is an identity matrix of the appropriate
dimension. Taking the full sample, create a plot of the condition number of the
covariance matrix versus the shrinkage parameter. How much shrinkage is necessary
to achieve a well-conditioned covariance matrix? For the rest of the exercises below,
set shrinkage to 0.5.

```python
def calc_covariance_matrix(df, date_index, shrink):
    sub_df = df.iloc[:date_index]
    S = np.diag(np.sqrt(np.array(sub_df.var())))
    R = np.array(sub_df.corr())    
    R_shrink = shrink*R + (1-shrink)*np.identity(num_countries)
    Sigma = np.dot(np.dot(S,R_shrink),S)
    cond_num = np.linalg.cond(Sigma)
    return Sigma, cond_num
```

```python
shrinks = np.arange(0,1.01,0.01)
num_rows = len(totret_df)
cond_num_list = []
for shrink in shrinks:
    Sigma, cond_num = calc_covariance_matrix(totret_df, num_rows, shrink)
    cond_num_list.append(cond_num)
```

```python
plt.plot(shrinks, cond_num_list)
plt.xlabel('Shrinkage')
plt.title('Condition Number of Shrinked Covariance Matrix')
plt.grid(True)
plt.show()
```

#### Problem 4.4

Write a function which takes an input data frame and computes the Black-Litterman portfolio appropriate for use in the last month of the input. Use the views computed in the previous part, and use parameter values $\tau = 0.01$, $\kappa = 1.0$. Compute the Black–Litterman portfolio $h^*$, print out its weights, and also create a
scatterplot with $h_{\text{eq}}$ weights on the x-axis and $h^∗$ weights on the y-axis.

**Hints**: Let $t$ be the date of the last row in the input data. Your program should
first compute the covariance matrix $\Sigma$ using the function you wrote earlier, but
be careful about what you pass as input to that function. Specifically, be sure to
use only returns up to and including $t−1$ in order to avoid lookahead bias. Then
compute the CAPM equilibrium $h_{\text{eq}}$ using the provided market capitalizations, and normalize so that $\sum_{i}h_{\text{eq},i}=1$.

```python
kappa = 1.0
tau = 0.01
q = 0.01
omega = 0.015
Omega = omega*omega
```

```python
# The correct way to add two covariance matrix:
# r~N(Theta, S),
# theta~N(Pi, C),
# then, Var(r)=(S^-1-S^-1 H^-1 S^-1)^-1,
# where H = S^-1 + C^-1.
def convolution_add(C, S):
    S_inv = np.linalg.inv(S)
    C_inv = np.linalg.inv(C)
    H_inv = np.linalg.inv(S_inv+C_inv)
    return np.linalg.inv(S_inv - np.dot(np.dot(S_inv, H_inv),S_inv))
```

```python
def calc_black_litterman_portfolio(totret_df, mktcap_df, date_index, tau, kappa, omega, q, how_to_add_cov='simple'):
    
    mktcap = np.array(mktcap_df.iloc[date_index])
    totret = np.array(totret_df.iloc[date_index])
    Omega = omega*omega
    
    shrink = 0.5
    Sigma, cond_num = calc_covariance_matrix(totret_df, date_index, shrink)    
    C = tau*Sigma
    
    # equilibrium
    h_eq = mktcap/np.sum(mktcap)
    pi = kappa*(1+tau)*np.dot(Sigma, h_eq)
    
    # view
    X_weights = np.zeros(num_countries)
    X_weights[DNK_INDEX] = 0.25
    X_weights[FIN_INDEX] = 0.25
    X_weights[NOR_INDEX] = 0.25
    X_weights[SWE_INDEX] = 0.25
    
    Y_weights = mktcap
    Y_weights[DNK_INDEX] = 0
    Y_weights[FIN_INDEX] = 0
    Y_weights[NOR_INDEX] = 0
    Y_weights[SWE_INDEX] = 0
    Y_weights = Y_weights/np.sum(Y_weights)
    
    np.array([X_weights - Y_weights])
    
    H_inv = np.linalg.inv(np.dot(np.dot(P.T, 1.0/Omega), P) + np.linalg.inv(C))
    eta = np.dot(np.dot(P.T, 1.0/Omega),q) + np.array([np.dot(np.linalg.inv(C), pi)]).T
    E_theta = np.dot(H_inv, eta)
    
    Vr = None
    if how_to_add_cov == 'simple':
        Vr = H_inv + Sigma
    elif how_to_add_cov == 'convolution':
        Vr = convolution_add(Sigma, H_inv)
        
    if Vr is None:
        print("Error: invalid covariance adding method specified!")
        return
    
    Vr_inv = np.linalg.inv(Vr)
    h_star = (1.0/kappa)*np.dot(Vr_inv, E_theta)
    h_star = h_star.T[0]
    
    return h_star, h_eq, totret_df.index.values[date_index], np.dot(h_star, totret), np.dot(h_eq, totret)
```

```python
h_star,h_eq,_,_,_ = calc_black_litterman_portfolio(totret_df, mktcap_df, len(totret_df)-1, tau, kappa, omega, q, how_to_add_cov='simple')
```

```python
position_df = pd.DataFrame()
position_df['Country'] = ['aut','bel','che','deu','dnk','esp','fin','fra','gbr','irl','ita','nld','nor','prt','swe']
position_df['h_eq'] = h_eq
position_df['h*'] = h_star
position_df.set_index('Country',inplace=True)
position_df
```

#### Problem 4.5

Using the function you wrote in the previous part, write a simple
backtest which, for each $t$, computes the BL optimal portfolio $h^*$ and equilibrium
portfolio $h_{\text{eq}}$ that were knowable as of the last trading day of month $t−1$, and
compute the dot product of these portfolios with the returns in month $t$. Plot the
cumulative sum of the two time-series of returns $h ^ * _ {t−1} \cdot r_t$ and $h _ {\text{eq},t−1} \cdot r_t$ on the same axes. Start from $t = \lfloor{N/4}\rfloor$ where $N$ is the total number of dates (after all,
you need some history to compute the covariance matrix.)

```python
start_date_index = int(num_rows/4)
start_date_index
```

```python
import datetime

ret_star_list = []
ret_eq_list = []
date_list = []
for date_index in range(start_date_index, num_rows):
    h_star, h_eq, date_str, ret_star, ret_eq = calc_black_litterman_portfolio(totret_df, mktcap_df, date_index, tau, kappa, omega, q, how_to_add_cov='simple')
    ret_star_list.append(ret_star)
    ret_eq_list.append(ret_eq)
    date_obj = datetime.datetime.strptime(date_str, '%Y-%m-%d')
    date_list.append(date_obj)
```

```python
cum_ret_star = np.cumsum(ret_star_list)
cum_ret_eq = np.cumsum(ret_eq_list)

plt.plot(date_list, cum_ret_star, label='$h^*$')
plt.plot(date_list, cum_ret_eq, label='$h_{eq}$')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.legend(bbox_to_anchor=(0.05,0.95), loc="upper left")
plt.grid(True)
plt.show()
```
